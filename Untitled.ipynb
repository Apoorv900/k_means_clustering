```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("Mall_Customers.csv")
X=data.iloc[:,[3,4]].values
```

# Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5 , init = 'k-means++' , random_state=42)
y_kmeans=kmeans.fit_predict(X)
print(y_kmeans)
```

# Visualising the clusters

```python
for i in range(5):
    plt.scatter(X[y_kmeans==i,0],X[y_kmeans==i,1],s=50,c=['red','green','blue','pink','black'][i],label="cluster"+str(i+1))
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=250,c="yellow",label='centroids')
plt.title("cluster by K means")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

```python
for i in range(6):
    plt.scatter(X[y_kmeans==i,0],X[y_kmeans==i,1],s=80,color=['red','blue','black','pink','green','orange'][i],
                label="cluster"+str(i+1))
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c="yellow",s=130,label="centroids")
plt.title("kmeans with 6 clusters")
plt.xlabel("Annual income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
    
```
